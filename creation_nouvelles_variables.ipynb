```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_row",50)
pd.set_option("display.max_columns",130)
```

```python
data_presidentielle_premier_tour_orig = pd.read_csv("C:/Users/Christian/Desktop/IRIS_BV_final/PRESIDENTIELLES/IRIS_FINAL_POURCENTAGE_1_TOUR.txt",sep=";", header=0, dtype=str)
data_presidentielle_premier_tour = data_presidentielle_premier_tour_orig.copy()

data_presidentielle_second_tour_orig = pd.read_csv("C:/Users/Christian/Desktop/IRIS_BV_final/PRESIDENTIELLES/SECOND_TOUR/IRIS_FINAL_POURCENTAGE_2_TOUR.txt",sep=";", header=0, dtype=str)
data_presidentielle_second_tour = data_presidentielle_second_tour_orig.copy()

data_europeen_orig=  pd.read_csv("C:/Users/Christian/Desktop/IRIS_BV_final/EUROPEEN/IRIS_FINAL_POURCENTAGE_EUROPEEN.txt",sep=";", header=0, dtype=str)
data_europeen = data_europeen_orig.copy()

```

```python
# COMPLETER LES NAN PAR DES 0
data_presidentielle_premier_tour = data_presidentielle_premier_tour.fillna("0")
data_presidentielle_second_tour =  data_presidentielle_second_tour.fillna("0")
data_europeen = data_europeen.fillna("0")


```

```python
data_europeen_6_quantitatives = data_europeen[["Voix/Exprimés_1","Voix/Exprimés_2","Voix/Exprimés_3","Voix/Exprimés_4","Voix/Exprimés_9","Voix/Exprimés_11"]]
data_presidentielle_premier_tour_6_quantitatives = data_presidentielle_premier_tour[["Voix/Exprimés_1","Voix/Exprimés_2","Voix/Exprimés_3","Voix/Exprimés_4","Voix/Exprimés_9","Voix/Exprimés_11"]]
data_presidentielle_second_tour_6_quantitatives = data_presidentielle_second_tour[["Voix/Exprimés_EM","Voix/Exprimés_MLP"]]
```

```python
# CONVERSION DES VARIABLES QUANTITATIVES EN QUALITATIVES
from sklearn.preprocessing import KBinsDiscretizer

```

```python
# PAS DE PROBLEME AVEC LES INDEX CAR ON VIENT D'APPELER LES FICHIERS
def conversion(data_temp, liste_de_col):
    method_discre = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='quantile') 
    method_discre.fit(data_temp)
    resultat_discre = method_discre.transform(data_temp)
    

    # construction du data frame après discrétisation
    df_resultat_discre = pd.DataFrame(resultat_discre,columns=liste_de_col, dtype=int )
    data_final = pd.concat([data_temp, df_resultat_discre], axis=1 ) #data2_index
    return data_final
def ajoute(data_1,data_2):
    data = pd.concat([data_1,data_2], axis=1)
    return data
```

```python
data_europeen_6_tous = conversion(data_europeen_6_quantitatives,["V_1","V_2","V_3","V_4","V_9","V_11"])
data_premier_tour_6_tous = conversion(data_presidentielle_premier_tour_6_quantitatives,["V_1","V_2","V_3","V_4","V_9","V_11"])
data_second_tour_6_tous = conversion(data_presidentielle_second_tour_6_quantitatives,["V_EM","V_MLP"])

```

```python
data_europeen_nouvelles_variables = ajoute(data_europeen ,data_europeen_6_tous.drop(["Voix/Exprimés_1","Voix/Exprimés_2","Voix/Exprimés_3","Voix/Exprimés_4","Voix/Exprimés_9","Voix/Exprimés_11"], axis=1))
data_premier_tour_nouvelles_variables = ajoute(data_presidentielle_premier_tour ,data_premier_tour_6_tous.drop(["Voix/Exprimés_1","Voix/Exprimés_2","Voix/Exprimés_3","Voix/Exprimés_4","Voix/Exprimés_9","Voix/Exprimés_11"], axis=1))
data_second_tour_nouvelles_variables =ajoute(data_presidentielle_second_tour, data_second_tour_6_tous.drop(["Voix/Exprimés_EM","Voix/Exprimés_MLP"],axis=1))
```
